```python
import numpy
from numpy import pi, sin, cos, sqrt, abs, log, power
import matplotlib.pyplot as plot
```

```python
def polar1():
    t = numpy.linspace(0, pi * 2, 100)
    r = 1 - sin(t)
    plot.polar(t, r)
    
polar1()
```

```python
def polar2():
    t = numpy.linspace(0, pi * 2, 100)
    r = (sin(t) * sqrt(abs(cos(t)))) / (sin(t) + 7 / 5) - 2 * sin(t) + 2
    plot.polar(t, r)
    
polar2()    
```

```python
def para1():
    t = numpy.linspace(-1, 1, 100)
    x = sin(t) * cos(t) * log(abs(t))
    y = power(abs(t), 0.3) * sqrt(cos(t))
    plot.plot(x, y ** 3)
    
para1()
```

```python
def para2():
    t = numpy.linspace(-pi * 2, pi * 2, 100)
    x = 16 * sin(t) ** 3
    y = 13 * cos(t) - 5 * cos(2 * t) - 2 * cos(3 * t) - cos(4 * t)
    plot.plot(x, y)

para2()
```

```python
def impl1():
    x = numpy.linspace(-2, 2, 100)
    y = numpy.linspace(-2, 2, 100)
    X, Y = numpy.meshgrid(x, y)
    F = (X ** 2 + Y ** 2 - 1) ** 3 - X ** 2 * Y ** 3
    plot.contour(X, Y, F, [0])
    
impl1()    
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots


def impl1_3d():
    x = numpy.linspace(-2, 2, 100)
    y = numpy.linspace(-2, 2, 100)
    X, Y = numpy.meshgrid(x, y)
    Z = (X ** 2 + Y ** 2 - 1) ** 3 - X ** 2 * Y ** 3
    W = X * 0 + Y * 0
    canvas = make_subplots(rows=1, cols=1, specs=[[{'type': 'surface'}]])
    surface = go.Surface(x=X, y=Y, z=W, colorscale="Hot")
    surface0 = go.Surface(x=X, y=Y, z=Z, colorscale="Greys")
    canvas.add_trace(surface, row=1, col=1)
    canvas.add_trace(surface0, row=1, col=1)
    canvas.show()


impl1_3d()
```

```python
def impl2():
    x = numpy.linspace(-8, 8, 100)
    y = numpy.linspace(-12, 6, 100)
    X, Y = numpy.meshgrid(x, y)
    F = (Y - (2 * (abs(X) + X ** 2 - 6)) / (3 * abs(X) + X ** 2 + 2)) ** 2 + X ** 2
    plot.contour(X, Y, F, [36])
    
impl2()
```
